```python
Question 3(A) : Compute Total Impedance of an Electrical Circuit
Problem Statement:
Write a Python program to calculate the total impedance of an electrical circuit using the
given impedance values of individual components—resistors, capacitors, and inductors.
Your program should be able to:
● Compute total impedance for components connected in series and parallel.
● Support impedance values in complex form (to account for frequency-dependent
elements like capacitors and inductors).
● Optionally allow the user to specify the circuit configuration and component values.
```

```python
def calculate_impedance(components, connection_type='series'):
    """
    Calculates total impedance for a list of complex impedance values.

    :param components: List of complex numbers (e.g., [10+0j, 0-5j])
    :param connection_type: 'series' or 'parallel'
    :return: Total complex impedance
    """
    if not components:
        return 0

    if connection_type.lower() == 'series':

        return sum(components)

    elif connection_type.lower() == 'parallel':

        reciprocal_sum = sum(1 / z for z in components if z != 0)
        return 1 / reciprocal_sum if reciprocal_sum != 0 else 0

    else:
        raise ValueError("Connection type must be 'series' or 'parallel'")




circuit_elements = [100 + 0j, 0 + 50j, 0 - 30j]


z_series = calculate_impedance(circuit_elements, 'series')
z_parallel = calculate_impedance(circuit_elements, 'parallel')

print(f"Components: {circuit_elements}")
print(f"Total Series Impedance: {z_series} Ohms")
print(f"Total Parallel Impedance: {z_parallel:.2f} Ohms")
print(f"Magnitude (Series): {abs(z_series):.2f} Ohms")
```
