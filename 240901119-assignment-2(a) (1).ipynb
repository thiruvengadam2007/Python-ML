```python
Write a Python program using NumPy to analyze your daily step count data collected over
the past month. The program should:
● Compute key statistics such as total steps, average steps per day, and
maximum/minimum step counts.
● Identify days with high activity (e.g., days with steps above a certain threshold).
● Analyze trends over time, such as increases or decreases in activity, and optionally
visualize them.
```

```python
import numpy as np


np.random.seed(42)
grades = np.random.randint(50, 101, (30, 3))

math_avg, sci_avg, eng_avg = np.mean(grades, axis=0)
math_max, sci_max, eng_max = np.max(grades, axis=0)

print(f"Math: Avg {math_avg:.1f}, High {math_max}")
print(f"Science: Avg {sci_avg:.1f}, High {sci_max}")
print(f"English: Avg {eng_avg:.1f}, High {eng_max}")


total_avg = np.mean(grades)
print(f"\nOverall Class Average: {total_avg:.1f}")

total_scores = np.sum(grades, axis=1)


best_score = np.max(total_scores)
best_student_id = np.argmax(total_scores)

print(f"Top Student is ID #{best_student_id} with {best_score} total points.")
```
