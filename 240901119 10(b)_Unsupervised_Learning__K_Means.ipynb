```python
Unsupervised Learning - K-Means

Aim:
To implement unsupervised learning-K-Means using any predefined dataset
Description:
K-means is an unsupervised learning method for clustering data points.
The algorithm iteratively divides data points into K clusters by minimizing
the variance in each cluster.

First, each data point is randomly assigned to one of the K clusters. Then, we
compute the centroid (functionally the center) of each cluster, and reassign each
data point to the cluster with the closest centroid. We repeat this process until
the cluster assignments for each data point are no longer changing.
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data = list(zip(x, y))

kmeans = KMeans(n_clusters=2, random_state=0) # Added random_state for reproducibility
kmeans.fit(data)

# Removed the '4/5' as it seemed like an incomplete line or comment

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
