```python
import pandas as pd
pd.set_option('display.max_columns',10)
print("Original data:\n")
df=pd.read_excel("CLEANING EXCEL.xlsx")
print(df)
```

```python
f=df['NAME'].str.strip()
f.to_excel("CLEANING EXCEL.xlsx")
print("\nReplace value:")
k=df.fillna('pad')
print(k)
print("\nUpdated list\n")
k.to_excel("CLEANING EXCEL.xlsx")
```

```python
#Updated list
print("\nDrop one particular column and its values:\n")
k.drop(['SOCIAL'], axis=1,inplace=True)
print(k)
print("\nUpdated list..\n")
k.to_excel("CLEANING EXCEL.xlsx")
```

```python
#Updated list..
print("\nDrop NaN rows:\n")
df=pd.read_excel("CLEANING EXCEL.xlsx")
print(df)
x=df.dropna()
print(x)
print("\nUpdated data list:\n")
```

```python
#Updated data set
print("\nReplace values:\n")
n = pd.read_excel("CLEANING EXCEL.xlsx") # Corrected npd to pd and assigned to n
print("\nOriginal data:\n")
print(n)
y = n.replace({89:883}) # Added assignment operator =
print("\nReplaced dataset:\n")
print(y)
print("\nUpdated dataset:\n")
y.to_excel("CLEANING EXCEL.xlsx")
```
