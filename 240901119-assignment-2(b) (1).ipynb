```python
Problem Statement:
Write a Python program using NumPy to analyze the grades of 30 students across three
subjects: Math, Science, and English. Your program should:
● Calculate overall and subject-wise statistics such as average, highest, and lowest
grades.
● Identify top-performing students based on total or average scores.
● Optionally visualize the grade distributions for each subject using histograms or bar
charts.
```

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
grades = np.random.randint(50, 101, size=(30, 3))
subjects = ['Math', 'Science', 'English']


subj_avg = np.mean(grades, axis=0)
subj_max = np.max(grades, axis=0)
subj_min = np.min(grades, axis=0)


overall_avg = np.mean(grades)


student_totals = np.sum(grades, axis=1)
top_student_idx = np.argmax(student_totals)
top_score = student_totals[top_student_idx]


print(f"Overall Class Average: {overall_avg:.2f}")
print("-" * 30)
for i, subj in enumerate(subjects):
    print(f"{subj}: Avg={subj_avg[i]:.1f}, High={subj_max[i]}, Low={subj_min[i]}")

print("-" * 30)
print(f"Top Student is Student #{top_student_idx} with a total of {top_score}")


plt.figure(figsize=(10, 5))
plt.boxplot(grades, tick_labels=subjects)
plt.title('Grade Distribution per Subject')
plt.ylabel('Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
