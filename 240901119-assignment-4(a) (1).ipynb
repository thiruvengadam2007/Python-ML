```python
Question 4 (A) : Analyze EEG Data Using Pandas
Problem Statement:
You have collected EEG data from multiple participants while they performed a cognitive
task. Each dataset includes time-series measurements of brainwave activity across various
electrodes. Write a Python program using Pandas to:
● Load and organize the EEG data.
● Perform basic statistical analysis (mean, variance, peak values) for each participant
and electrode.
● Identify patterns or trends in brainwave activity during the task.
```

```python
import pandas as pd
import numpy as np

data = {
    'Participant_ID': np.repeat(['P1', 'P2'], 50),
    'Timestamp': np.tile(np.arange(0, 5, 0.1), 2), # 5 seconds of data
    'Fp1': np.random.normal(20, 5, 100), # Frontal (Attention)
    'Cz': np.random.normal(15, 3, 100),  # Central (Motor/Sensor)
    'O1': np.random.normal(10, 2, 100)   # Occipital (Visual)
}
df = pd.DataFrame(data)

def analyze_eeg_live(df):
    electrodes = ['Fp1', 'Cz', 'O1']

    stats = df.groupby('Participant_ID')[electrodes].agg(['mean', 'var', 'max'])

    print("### 1. Statistical Summary ###")
    print(stats)
    print("\n" + "-"*30 + "\n")

    print("### 2. Peak Activity Trends ###")
    for electrode in electrodes:
        peak_row = df.loc[df[electrode].idxmax()]
        print(f"Highest activity in {electrode}: {peak_row[electrode]:.2f} units at {peak_row['Timestamp']}s")

    return stats

results = analyze_eeg_live(df)
```
