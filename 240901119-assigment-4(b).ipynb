```python
Question4 (B): Perform Basic Data Analysis on Product Inventory Using
Pandas
Problem Statement:
You are given a dictionary containing information about products in an inventory, including
their names, prices, and quantities available:
data = {'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],'Price': [1200, 800,
300, 150, 50],'Quantity': [10, 20, 15, 25, 50]} and do any five basic operations and plot.
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 300, 150, 50],
    'Quantity': [10, 20, 15, 25, 50]
}

df = pd.DataFrame(data)




df['Total_Value'] = df['Price'] * df['Quantity']
stats = df.describe()


expensive_products = df[df['Price'] > 200]


sorted_df = df.sort_values(by='Quantity', ascending=False)


top_product = df.loc[df['Total_Value'].idxmax()]


print("--- Full Inventory DataFrame ---")
print(df)
print("\n--- Descriptive Statistics ---")
print(stats)
print("\n--- Products Price > 200 ---")
print(expensive_products)
print("\n--- Top Product by Inventory Value ---")
print(top_product)


plt.figure(figsize=(10, 6))
plt.bar(df['Product'], df['Total_Value'], color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Total Inventory Value ($)')
plt.title('Total Inventory Value per Product')
plt.show()
```
