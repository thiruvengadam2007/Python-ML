```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px

# seaborn style
sns.set()

# Data (same length)
Xpoints = np.array([0, 20, 40, 80, 120, 160, 200])
Ypoints = np.array([10, 80, 100, 120, 150, 180, 250])

# Line plot
plt.plot(Xpoints, Ypoints)

plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
print(sns.scatterplot(data=Ypoints))
print(plt.show())
```

```python
print(sns.barplot(data=Ypoints))
print(plt.show())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

# ---------- 3D Scatter Plot ----------
X = np.array([1, 2, 6, 8, 12, 14, 20])
Y = np.array([3, 8, 1, 10, 12, 16, 18])
Z = np.array([0.01, 0.02, 0.00, 0.03, -0.01, 0.02, 0.00])  # small values like image

fig = plt.figure(figsize=(10, 7))
ax = plt.axes(projection='3d')

plot_3d = ax.scatter3D(X, Y, Z, c=Z, cmap='viridis')  # color mapping

fig.colorbar(plot_3d)  # color bar like your image

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.title("3D Scatter Plot")
plt.show()


# ---------- Plotly Scatter ----------
X = np.array([10, 80, 100, 120, 150, 180, 250])
Y = np.array([1, 2, 3, 4, 5, 6, 7])

fig = px.scatter(x=X, y=Y, title="Plotly Scatter Plot")
fig.show()
```
