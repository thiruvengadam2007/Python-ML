```python
  PROBLEM STATEMENT:
    Write a python program to solve a system of linear equations represented by:
      Ax=b
  Where:
     A is a  2*2 matrix
     B is a 2*1 matrix
     The program should  compute and return the solution vector x
```

```python
import numpy as np

def solve_linear_system(A_matrix, b_vector):
    """
    Solves the system Ax = b using NumPy.
    """
    try:
        # Convert inputs to numpy arrays
        A = np.array(A_matrix)
        b = np.array(b_vector)

        # Solve for x
        x = np.linalg.solve(A, b)
        return x

    except np.linalg.LinAlgError:
        return "Error: Matrix A is singular and cannot be solved."


A = [[3, 1],
     [1, 2]]

b = [9, 8]

solution = solve_linear_system(A, b)

print(f"Matrix A:\n{np.array(A)}")
print(f"Vector b: {b}")
print(f"Solution vector x: {solution}")
```
